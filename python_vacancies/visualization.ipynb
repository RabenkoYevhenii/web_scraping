```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("vacancies.csv")
```

```python
technologies_df = df["Technologies"].str.split(",", expand=True)
```

```python
technology_counts = technologies_df.apply(pd.Series.value_counts).sum(axis=1)
```

```python
sorted_technology_counts = technology_counts.sort_values(ascending=False)
```

```python
plt.figure(figsize=(12, 8))
sorted_technology_counts.plot(kind="bar", color="skyblue")
plt.title("Popularity of Technologies")
plt.xlabel("Technologies")
plt.ylabel("Number of Occurrences")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
english_level_counts = df["English level"].value_counts()
```

```python
plt.figure(figsize=(10, 6))
english_level_counts.plot(kind="bar", color="skyblue")
plt.title("Distribution of English Levels")
plt.xlabel("English Levels")
plt.ylabel("Number of Occurrences")
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
